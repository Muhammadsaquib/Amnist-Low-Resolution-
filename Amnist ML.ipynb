### Spliting the data in train and test

```python
(train_images,train_labels), (test_images,test_labels) = data.load_data()
```

### Designing NN layers
#### first one is flatten layer which convert data in one list
#### Second is hidden layer ,size of neurons and with that activation func
#### Third is the output layer , size of neurons and activation func
#### (Softmax func : it's convert all output in sum of 1 in total

```python
model  = keras.Sequential([
        keras.layers.Flatten(input_shape = (28,28)),
        keras.layers.Dense(128, activation= "relu"),
        keras.layers.Dense(10,activation = "softmax")

])
```

### Configures the model for training (Compile function)

```python
model.compile(optimizer ="adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

#### epoch is to showing your model the whole training data bunch once
#### (Here is 10 means 10 times showing you training data to model for train it)

```python
model.fit(train_images, train_labels, epochs = 10)
```

```python
test_loss,test_acc = model.evaluate( test_images,test_labels)
print("test acc :", test_acc)
```

```python
prediction = model.predict(test_images)
```
